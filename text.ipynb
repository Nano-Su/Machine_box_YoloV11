```python
import cv2
from paddleocr import PaddleOCR
from pprint import pprint

```

```python
img_path = 'img.png'
ocr = PaddleOCR(lang='ch',device='gpu',use_angle_cls=True)
```

```python
result = ocr.predict(img_path)
```

```python
# 假设 predict_result 是你得到的那个复杂的列表
# 我们只处理第一张图片的结果
res =result[0]

texts = res['rec_texts']
scores = res['rec_scores']
boxes = res['rec_polys']

#文字信息
for i in range(len(texts)):
    print(f"文字: {texts[i]}")
    print(f"  - 置信度: {scores[i]:.4f}")
    print("-" * 20)
#框选列表
for i in range(len(boxes)):
    print(f"框选坐标: {boxes[i]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 读取图像
image = cv2.imread(img_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 绘制框选
for box in boxes:
    box = np.array(box).astype(np.int32).reshape((-1, 1, 2))
    cv2.polylines(image, [box], True, (0, 255, 0), 2)

# 显示图像
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis('off')
plt.show()
```
